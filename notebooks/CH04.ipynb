```python
import tensorflow as tf
x = tf.placeholder(tf.float32, (3,1))
y = tf.placeholder(tf.float32, (3,1))
xT = tf.transpose(x)
inner_product = tf.matmul(xT, y)
with tf.Session() as sess:
    out = sess.run(inner_product, feed_dict={ x : [[1], [2], [3]], y : [[4], [5], [6]]})
    print("{0}".format(out))
```

```python
X = tf.placeholder(tf.float32, (5,1))
w = tf.Variable([[2.0]])
b = tf.Variable(1.0)
y = tf.matmul(X,w) + b

init = tf. global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    data, out = sess.run([X, y], feed_dict={ X : [[-1.0], [-0.1], [0.9], [2.0], [3.1]]})
    print("{0}".format(out))
    print("X의 Shape : {0}".format(X.shape))
    print("w의 Shape : {0}".format(w.shape))
    print("b의 Shape : {0}".format(b.shape))
    print("y의 Shape : {0}".format(out.shape))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

plt.plot(data, out, 'o:')
plt.grid()
```

```python
X = tf.placeholder(tf.float32, (5,2))
w = tf.Variable([[2.0], [1.0]])
b = tf.Variable(1.0)
y = tf.matmul(X,w) + b

init = tf. global_variables_initializer ()
with tf.Session() as sess:
    sess.run(init)
    data, out = sess.run([X, y], feed_dict={ X : [[-1.0, 1.0], [-0.1, 0.3], [0.9, 1.2], [2.0, 2.0], [3.1, 2.9]]})
    print("{0}".format(out))
    print("X의 Shape : {0}".format(X.shape))
    print("w의 Shape : {0}".format(w.shape))
    print("b의 Shape : {0}".format(b.shape))
    print("y의 Shape : {0}".format(out.shape))
```
